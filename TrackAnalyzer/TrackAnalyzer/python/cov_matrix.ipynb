```python
covMatrixLab=["theTrack_SQopQop",
"theTrack_SQopLam",
"theTrack_SQopPhi",
"theTrack_SQopDxy",
"theTrack_SQopDsz",
"theTrack_SLamLam",
"theTrack_SLamPhi",
"theTrack_SLamDxy",
"theTrack_SLamDsz",
"theTrack_SPhiPhi",
"theTrack_SPhiDxy",
"theTrack_SPhiDsz",
"theTrack_SDxyDxy",
"theTrack_SDxyDsz",
"theTrack_SDszDsz"]
```

```python
def symmetricize(a):
    s = int((np.sqrt(a.size*8+1)-1)*0.5)
    m = np.zeros((s,s))
    for i in range(s):
        F = s + np.sum(np.arange(s-1,s-i-1,-1))
        I = F - s + i
        m[i,i:] = a[I:F]
    for i in range(s):
        m[i:,i] = m[i,i:]
    return m
```

```python
C_aod_vec = data_aod_selcta[covMatrixLab].values[:500000]
```

```python
C_mini_vec = data_mini_selcta[covMatrixLab].values[:500000]
```

```python
M_mini = np.array([symmetricize(f) for f in tqdm(C_mini_vec)])
V_mini = np.array([np.linalg.eigvals(M) for M in tqdm(M_mini)])
```

```python
M_aod = np.array([symmetricize(f) for f in tqdm(C_aod_vec)])
E_aod = np.array([np.linalg.eigvals(M) for M in tqdm(M_aod)])
```

```python
DISC_VARS = ["theTrack_NPixelHits","theTrack_NStripHits","theTrack_NTrackhits","theTrack_NBPixHits",
            "theTrack_NPixLayers","theTrack_NTraLayers","theTrack_NStrLayers","theTrack_NBPixLayers"]
CONT_VARS = ["theTrack_phi","theTrack_pt"]
```

```python
matches = []
for index, row in tqdm(bad_aod_data.iterrows()):
    df = bad_mini_data[bad_mini_data["event"]==row["event"]]

    if len(df)<1:
        continue
    for v in DISC_VARS:
        df = df[df[v].values==row[v]]
        if len(df)<1:
            break
    if len(df)<1:
        continue
    for v in DISC_VARS:
        df = df[(abs(df[v].values-row[v])/row[v])<0.01]
        if len(df)<1:
            break
    if len(df)<1:
        continue
    
    matches.append([index,df.index[0]])
```
